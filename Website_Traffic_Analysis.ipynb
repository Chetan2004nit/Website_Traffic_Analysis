```python
 import numpy as np
 import pandas as pd
 import matplotlib.pyplot as plt
 import seaborn as sns 
```

```python
 df = pd.read_csv("data-export (1).csv")
 df.head()
```

```python
df.columns = df.iloc[0]
df = df.drop(index = 0).reset_index(drop = True)
df.columns = [
    "channel group", 
    "DateHour", 
    "Users", 
    "Sessions", 
    "Engaged Sessions", 
    "Average engagement time per session",
    "Engaged sessions per user",
    "Events per session",
    "Engagement rate",
    "Event count"
]
```

```python
df.head()
```

```python
df.info()
```

## cleaning data and data validation for the above data set

```python
df.head() 
```

```python
df["DateHour"] = pd.to_datetime(df["DateHour"], format="%Y%m%d%H", errors='coerce')
```

```python
numeric_cols = df.columns.drop(["channel group", "DateHour"])  # Dropping non-numeric columns to select only numeric ones
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')  # Converting all selected columns to numeric, coercing errors to NaN
df["Hour"] = df["DateHour"].dt.hour  # Extracting the hour part from the datetime column
```

# sessions and user over time

```python
 sns.set(style="whitegrid")
```

```python
 plt.figure(figsize=(10,5))
 df.groupby("DateHour")[["Sessions","Users"]].sum().plot(ax=plt.gca())
 plt.title("Sessions and users over time")
 plt.xlabel("DateHour")
 plt.ylabel("count")
 plt.show()
```

# total users by channel

```python

 plt.figure(figsize=(8, 5))
 sns.barplot(data=df, x="channel group", y="Users", estimator=np.sum, palette="viridis")
 plt.title("👥Total Users by Channel")
 plt.xticks(rotation=45)
 plt.show()
```

```python
df.head()
```

# Average engagement time by channel

```python
plt.figure(figsize=(8, 5))  # Set the size of the figure
sns.barplot(data=df, x="channel group", y="Average engagement time per session", estimator=np.mean, palette="crest")  # Plot average engagement time per channel
plt.title("Avg Engagement Time by Channel")  # Title for the chart
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

# Engagement Rate Distribution by channel

```python
 plt.figure(figsize=(8, 5))
 sns.boxplot(data=df, x="channel group", y="Engagement rate", palette="coolwarm")
 plt.title("Engagement Rate Distribution by Channel")
 plt.xticks(rotation=45)
 plt.show()
```

# Engaged vs non engaged sessions

```python
df.head()
```

```python
# Grouping by channel group and summing session columns
session_df = df.groupby("channel group")[["Sessions", "Engaged Sessions"]].sum()

# Creating a new column for non-engaged sessions
session_df["Non-Engaged"] = session_df["Sessions"] - session_df["Engaged Sessions"]

# Melting the DataFrame to long format for grouped barplot
session_df_melted = session_df.reset_index().melt(
    id_vars="channel group",
    value_vars=["Engaged Sessions", "Non-Engaged"],
    var_name="Session Type",
    value_name="value"
)

# Plotting the grouped bar chart
plt.figure(figsize=(8, 5))
sns.barplot(data=session_df_melted, x="channel group", y="value", hue="Session Type", palette="Set2")
plt.title("🧍‍♂️ Engaged vs Non-Engaged Sessions")  # Title with emoji
plt.xticks(rotation=45)  # Rotating x-axis labels
plt.tight_layout()  # Adjust layout to prevent label overlap
plt.show()
```

# traffic by hour and channel

```python
 heatmap_data = df.groupby(["Hour", "channel group"])["Sessions"].sum().unstack()
 plt.figure(figsize=(12, 6))
 sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=.5, annot=True, fmt='.0f')
 plt.title("⏰Traffic by Hour and Channel")
 plt.xlabel("Channel Group")
 plt.ylabel("Hour of Day")
 plt.show()
```

# engagement rate vs sessions over time

```python
df.head()
```

```python
# Grouping and aggregating the data
df_plot = df.groupby("DateHour")[["Engagement rate", "Sessions"]].mean().reset_index()

# Plotting
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df_plot["DateHour"], df_plot["Engagement rate"], label="Engagement rate", color="orange")
plt.plot(df_plot["DateHour"], df_plot["Sessions"], label="Sessions", color="blue")
plt.title("📊 Engagement Rate vs Sessions Over Time")
plt.xlabel("DateHour")
plt.ylabel("Values")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
